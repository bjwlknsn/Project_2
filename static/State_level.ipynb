# Import Dependencies

```python
#Import Necessary Dependencies 
import pandas as pd
import os
import matplotlib.pyplot as plt
```

# Import File Paths and Create DFs

```python
#File Paths
brewery_path=os.path.join('..','Resources','brewery_count.csv')
beer_pub_path=os.path.join('..','Resources','taxable_barrels_pubs_breweries.csv')
beer_can_path=os.path.join('..','Resources','taxable_barrels_bottles_cans.csv')
beer_keg_path=os.path.join('..','Resources','taxable_barrels_barrels_kegs.csv')

#Data Frames
brewery_0_df=pd.read_csv(brewery_path)
beer_pub_0_df=pd.read_csv(beer_pub_path)
beer_can_0_df=pd.read_csv(beer_can_path)
beer_keg_0_df=pd.read_csv(beer_keg_path)

#Output File Path
output_path=os.path.join('..','Output')
```

# Clean Up Raw Data and Create Data Frame

```python
#Include only data from 2008 to 2019 due to data constraints
# brewery_0_df
# beer_pub_0_df
beer_can_0_df
# beer_keg_0_df
```

```python
#Create new brewery data frame, filtering the brewery count data to only 2008 to 2019 
#brewery_1_df

brewery_1_df=brewery_0_df[['STATE','2008','2009','2010','2011','2012','2013','2014','2015','2016','2017','2018','2019']]
brewery_1_df
```

```python
#Unpivot the tacable beer data frame containing data from 2008-2019, create as new data frame brewery_2_df, 
#then clean up column names and save as brewery_3_df
brewery_2_df=brewery_1_df.set_index('STATE').unstack().reset_index(name='Brewery Count')

column_names={'level_0':'Year',
              'STATE': 'State',
              'Brewery Count':'Brewery Count'}

brewery_2_rename_df=brewery_2_df.rename(columns=column_names)
brewery_3_df=brewery_2_rename_df[['State','Year','Brewery Count']]
brewery_3_df
```

```python
#Unpivot the beer sold at pubs and breweries data frame, create as new data frame beer_pub_1_df, 
#clean up and organize columns as beer_pub_2_df

beer_pub_1_df=beer_pub_0_df.set_index('STATE').unstack().reset_index(name='Brewery and Pub Beer Sales (bbls)')
# beer_pub_1_df

column_names={'level_0':'Year',
              'STATE': 'State',
              'Brewery and Pub Beer Sales (bbls)':'Brewery and Pub Beer Sales (bbls)'}

beer_pub_1_rename_df=beer_pub_1_df.rename(columns=column_names)
beer_pub_2_df=beer_pub_1_rename_df[['State','Year','Brewery and Pub Beer Sales (bbls)']]
beer_pub_2_df
```

```python
#Unpivot the beer sold in cans or bottles data frame, create as new data frame beer_can_1_df, 
#clean up and organize columns as beer_can_2_df

beer_can_1_df=beer_can_0_df.set_index('State').unstack().reset_index(name='Can and Bottle Beer Sales (bbls)')
beer_can_1_df

column_names={'level_0':'Year',
              'State': 'State',
              'Can and Bottle Beer Sales (bbls)':'Can and Bottle Beer Sales (bbls)'}

beer_can_1_rename_df=beer_can_1_df.rename(columns=column_names)
beer_can_2_df=beer_can_1_rename_df[['State','Year','Can and Bottle Beer Sales (bbls)']]
beer_can_2_df
```

```python
# beer_keg_0_df

#Unpivot the beer sold in kegs and barrels data frame, create as new data frame beer_keg_1_df, 
#clean up and organize columns as beer_keg_2_df

beer_keg_1_df=beer_keg_0_df.set_index('State').unstack().reset_index(name='Keg and Barrel Beer Sales (bbls)')
beer_keg_1_df

column_names={'level_0':'Year',
              'State': 'State',
              'Keg and Barrel Beer Sales (bbls)':'Keg and Barrel Beer Sales (bbls)'}

beer_keg_1_rename_df=beer_keg_1_df.rename(columns=column_names)
beer_keg_2_df=beer_keg_1_rename_df[['State','Year','Keg and Barrel Beer Sales (bbls)']]
beer_keg_2_df
```

# Merge All Data Frames

```python
#Merging brewery_3_df and beer_pub_2_df into brewery_beer_0_df
brewery_beer_0_df=brewery_3_df.merge(beer_pub_2_df,on=['State','Year'],how='left')
# brewery_beer_0_df

#Merging brewery_beer_0_df and beer_can_2_df into brewery_beer_1_df
brewery_beer_1_df=brewery_beer_0_df.merge(beer_can_2_df,on=['State','Year'],how='left')
# brewery_beer_1_df

#Merging brewery_beer_1_df and beer_keg_2_df into brewery_beer_df
brewery_beer_2_df=brewery_beer_1_df.merge(beer_keg_2_df,on=['State','Year'],how='left')
# brewery_beer_2_df.count()

#Clean up beer and brewery data fram by dopping missing values
brewery_beer_3_df=brewery_beer_2_df.dropna(axis=0,how='any')
brewery_beer_3_df.count()

#Remove all commas from data frame
brewery_beer_4_df = brewery_beer_3_df.replace(',','', regex=True)

# Convert Brewery and Pub Beer Sales (bbls), Can and Bottle Beer Sales (bbls) and Keg and Barell Beer Sales (bbls) to strings

columns_dtypes={'Brewery and Pub Beer Sales (bbls)':'float', 
                'Can and Bottle Beer Sales (bbls)':'float',
                'Keg and Barrel Beer Sales (bbls)':'float'} 

brewery_beer_df=brewery_beer_4_df.astype(columns_dtypes)

#Include Total Beer Sales in Barrels within the Full Brewery and Beer Data Frame
brewery_beer_df['Total Beer Sales (bbls)']=(brewery_beer_df['Brewery and Pub Beer Sales (bbls)']+brewery_beer_df['Can and Bottle Beer Sales (bbls)']+brewery_beer_df['Keg and Barrel Beer Sales (bbls)'])
brewery_beer_df
```

```python
#Attmpt to pull in state level census data
# Dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import requests
from census import Census

# Census API Key
from config import api_key
c = Census(api_key, year=2013)
```

```python
# Run Census Search to retrieve data on all zip codes (2013 ACS5 Census)
# See: https://github.com/CommerceDataService/census-wrapper for library documentation
# See: https://gist.github.com/afhaque/60558290d6efd892351c4b64e5c01e9b for labels
census_data = c.acs5.get(("NAME", "B19013_001E", "B01003_001E", "B01002_001E",
                          "B19301_001E",
                          "B17001_002E"), {'for': 'zip code tabulation area:*'})

# Convert to DataFrame
census_pd = pd.DataFrame(census_data)

# Column Reordering
census_pd = census_pd.rename(columns={"B01003_001E": "Population",
                                      "B01002_001E": "Median Age",
                                      "B19013_001E": "Household Income",
                                      "B19301_001E": "Per Capita Income",
                                      "B17001_002E": "Poverty Count",
                                      "NAME": "Name", "zip code tabulation area": "Zipcode"})

# Add in Poverty Rate (Poverty Count / Population)
census_pd["Poverty Rate"] = 100 * \
    census_pd["Poverty Count"].astype(
        int) / census_pd["Population"].astype(int)

# Final DataFrame
census_pd = census_pd[["Zipcode", "Population", "Median Age", "Household Income",
                       "Per Capita Income", "Poverty Count", "Poverty Rate"]]

# Visualize
print(len(census_pd))
census_pd
```
